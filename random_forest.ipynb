```python
import pandas as pd
```

```python
data = pd.read_csv('rent_apartments.csv')
```

```python
data
```

### Data Prep

```python
cat_features = ['balcony', 'storage', 'parking', 'furnished', 'garage']
data_encoded = pd.get_dummies(data, columns=cat_features, drop_first=True)
```

```python
import re
```

```python
data_encoded.garden.unique()
```

```python
int(re.findall(r'\d+', data_encoded.garden[4])[0])
```

```python
for i in range(len(data_encoded)):
    if data_encoded.garden[i]=='Not present':
        data_encoded.garden[i]=0
    else: data_encoded.garden[i] = int(re.findall(r'\d+', data_encoded.garden[i])[0])
```

### Model Building

```python
IVs = ['area', 
       'constraction_year', 
       'bedrooms', 
       'garden', 
       'balcony_yes', 
       'parking_yes', 
       'furnished_yes', 
       'garage_yes', 
       'storage_yes'
      ]

X = data_encoded[IVs].copy()
dv = data_encoded.rent.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, dv, test_size=0.2)
```

### Model Selection

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

### Prediction

```python
rf.predict([[85, 2015, 2, 20, 1, 1, 0, 0, 1]])
```

### Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_space = {'n_estimators': [100, 200, 300], 'max_depth': [3, 6, 9, 12]}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid=grid_space, cv=5, scoring = 'r2')
```

```python
model_grid = grid.fit(X_train, y_train)
```

```python
print(f'Best hyperparameters are {model_grid.best_params_}, score = {model_grid.best_score_}')
```

```python
model_grid.best_estimator_.predict([[85, 2015, 2, 20, 1, 1, 0, 0, 1]])
```
